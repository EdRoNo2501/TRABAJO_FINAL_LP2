+ Acontinuación veremos como extraer información de la api con los parametros

```python
# Pauqetes necesarios
import requests # Requerimos de este paquete para hacer la solicitud de la api
import json


# Define la URL base de la API y tu clave de API
base_url = "https://api.nasa.gov/planetary/apod"
api_key = "2AxvQNHCDvyUedHwlhZg5bzViaLRWoz9qpOclGeI" #El enviado por el correo

# Realiza la solicitud HTTP GET
response = requests.get(f"{base_url}?api_key={api_key}") 
#En este caso al solo colocar el api key nos dara información solo del dia de hoy al hacer la solicitud

# Verifica si la solicitud fue exitosa
if response.status_code == 200:
    # La solicitud fue exitosa, extrae los datos
    data = response.json() # tiene forma de diccionario
    json_str = json.dumps(data, indent=4) # lo convierte en una cadena json
    print(json_str)# imprime mostrando los datos en formato JSON con sangría para una mejor legibilidad
else:
    # La solicitud falló, maneja el error apropiadamente
    print("Error al hacer la solicitud:", response.status_code)

```

+ Si queremos una cantidad en especifico y al azar usamos el parametro "count".

```python
# Pauqetes necesarios
import requests
import json

# URL de la API de APOD
url = "https://api.nasa.gov/planetary/apod"

# Parámetros de la solicitud
params = {
    "api_key": "2AxvQNHCDvyUedHwlhZg5bzViaLRWoz9qpOclGeI",
    "count": 5  # Cantidad de imágenes a obtener (en este ejemplo, 5)
}

# Realizar la solicitud GET a la API
response = requests.get(url, params=params)

# Verificar el código de estado de la respuesta
if response.status_code == 200:
    # Extraer los datos de la respuesta en formato JSON
    data = response.json()

    # Procesar los datos de nuestra data
    for item in data:
        date = item["date"]
        title = item["title"]
        explanation = item["explanation"]
        hd_url = item["hdurl"]
        
        # Una vez extraídos podemos imprimirlo de la siguiente forma:
        print("Fecha:", date)
        print("Título:", title)
        print("Explicación:", explanation)
        print("URL de alta resolución:", hd_url)
        print("---")
else: 
    print("Error al realizar la solicitud:", response.status_code)
```

+ Si queremos extraerlo en un periodo de tiempo determindo usamos los siguentes parametros.

```python
# Paquetes necesarios
import requests
import pandas as pd
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# PArametros para el rango que se quiere
start_date = "2023-06-01"  # Fecha de inicio del rango
end_date = "2023-06-05"  # Fecha de fin del rango

# Lista para almacenar los datos
apod_data = []

# Obtener la lista de fechas en el rango deseado
date_range = pd.date_range(start_date, end_date)

# Iterar sobre cada fecha en el rango
for date in date_range:
    formatted_date = date.strftime("%Y-%m-%d")
    url = f"https://api.nasa.gov/planetary/apod?api_key=2AxvQNHCDvyUedHwlhZg5bzViaLRWoz9qpOclGeI&date={formatted_date}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        apod_data.append(data)

# Procesar los datos obtenidos
for item in apod_data:
    date = item["date"]
    title = item["title"]
    explanation = item["explanation"]
    hd_url = item["hdurl"]
    # ... Hacer algo con los datos obtenidos ...
    print("Fecha:", date)
    print("Título:", title)
    print("Explicación:", explanation)
    print("URL de alta resolución:", hd_url)
    print("---")
```

+ En nuestra api nos puede proporcionar un video o imagen, por lo que en este codigo podras obtener la url del video o imagen. Si es una imagen se podra visualizar la imagen pero, si fuera un video se visualiza el link en el cual podras verlo y que es un video.

```python
#Paquetes requeridos:
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from IPython.display import HTML

# Hacer la solicitud a la API de APOD
api_key = '2AxvQNHCDvyUedHwlhZg5bzViaLRWoz9qpOclGeI'
url = f'https://api.nasa.gov/planetary/apod?api_key={api_key}'
response = requests.get(url)
data = response.json()

#Obtener el titulo de la imagen:
title = data["title"]
tipo_dato= data["media_type"]

# Verificamos el tipo de dato
if tipo_dato == "video": # si es un video
    video_url = data["url"]
    video_title = data["title"]
    video_html = f'<video src="{video_url}" controls></video>'
    display(HTML(f"<h2>{video_title}</h2><a href={video_url}> Enlace del video </a></p>{video_html}"))

elif tipo_dato == "image":
    # Obtener la URL de la imagen
    image_url = data['url']
    print("Título:", title)
    print("URL de la miniatura de la imagen:",image_url)

    # Obtener la imagen y mostrarla
    image_response = requests.get(image_url)
    image = Image.open(BytesIO(image_response.content))
    
    #Visualización de imagen
    plt.imshow(image)
    plt.axis('off')
    plt.show()
else:
    print("No se pudo cargar")
```

+ Se puede observar que en nuestra api no solo hay imagenes tambien videos, en este caso para poder visualizar cualquira de los 2 usamos el siguiente codigo 

```python
import requests
from IPython.display import Image, display
from IPython.display import Video

response = requests.get('https://api.nasa.gov/planetary/apod?api_key=2AxvQNHCDvyUedHwlhZg5bzViaLRWoz9qpOclGeI')

data = response.json()
url = data['url']
tipo_dato = data["media_type"] #con esto verificamos si es imagen o video

# Para la visualización de imagen o video
if tipo_dato == "image":
    display(Image(url=url)) #Para imagen
elif tipo_dato == "video":
    display(Video(url=url)) #Para video
else:
    print("No se pudo cargar la url")
```
