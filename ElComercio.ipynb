```python
import pandas as pd

class ClasificadorNoticias:
    def __init__(self, dataframe):
        self.dataframe = dataframe
    
    def clasificar_noticias(self, sentimiento):
        if sentimiento == "bueno":
            noticias_buenas = self.dataframe[self.dataframe['ANALISIS_DESCRIPCION'] >= 0.5] 
            return noticias_buenas[['TITULO', 'FECHA', 'URL']]
        elif sentimiento == "malo":
            noticias_malas = self.dataframe[self.dataframe['ANALISIS_DESCRIPCION'] <= 0.5]  
            return noticias_malas[['TITULO', 'FECHA', 'URL']]
        else:
            return "Sentimiento no válido"

# Crear una instancia de la clase ClasificadorNoticias  DOLAR
clasificador = ClasificadorNoticias(DOLAR)

sentimiento = input("Ingrese el sentimiento (bueno/malo): ")
noticias_clasificadas = clasificador.clasificar_noticias(sentimiento)

print(noticias_clasificadas)
```
