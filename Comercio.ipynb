```python
# Crear listas para almacenar los datos
titles = []
dates = []
references = []
links = []

translator = Translator()
base_url = "https://www.economist.com/search?q=dolar+2023&page={}"

for page in range(1, 7):  
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    results = soup.find_all("li", class_="_result-item")

    for result in results:
        title = result.find("span", class_="_headline").text.strip()
        date_and_reference = result.find("p", class_="_snippet").text.strip()
        link = result.find("a", class_="_search-result")["href"]

        # Verificar si la cadena contiene "..."
        if "..." in date_and_reference:
            # Extraer la fecha y referencia
            date, reference = date_and_reference.split("...", 1)
            date = date.strip()

            # Traducir el título y la referencia al español
            title = translator.translate(title, dest='es').text
            reference = translator.translate(reference, dest='es').text if reference else ""
        else:
            # Asignar la cadena completa a la fecha y establecer la referencia como vacía
            date = date_and_reference

        # Traducir el título al español
        title = translator.translate(title, dest='es').text
        
        # Agregar los datos a las listas
        titles.append(title)
        dates.append(date)
        references.append(reference.strip())
        links.append(link)


```

```python
# Crear un DataFrame
df = pd.DataFrame({"Titulo": titles, "Fecha": dates, "Referencia": references, "Enlace": links})

# Guardar como archivo CSV
df.to_csv("Articulos.The_Economist.csv", index=False)
```

```python
from textblob import TextBlob

# Definir una función para analizar el sentimiento en español
def analizar_sentimiento(texto):
    blob = TextBlob(texto)
    if blob.detect_language() != 'es':
        blob = blob.translate(to='es')
    return blob.sentiment.polarity

# Analizar el sentimiento en la columna "Título"
df['Sentimiento_Título'] = df['Título'].apply(analizar_sentimiento)

# Analizar el sentimiento en la columna "Cuerpo"
df['Sentimiento_Cuerpo'] = df['Cuerpo'].apply(analizar_sentimiento)
```
