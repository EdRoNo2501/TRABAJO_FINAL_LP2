# <h1 style="font-family:Impact"><b><span style="color:blue;">Propsito</span></b></h1>
El propsito de este proyecto es brindar a los compradores de viviendas en dlares una ventaja estratgica al tomar decisiones informadas y oportunas en un mercado inmobiliario globalizado y voltil. El sistema proporciona datos en tiempo real del tipo de cambio, obtenidos de fuentes confiables, y notifica a los usuarios cuando el tipo de cambio alcance niveles crticos establecidos por ellos mismos. En definitiva, el propsito es empoderar a los usuarios y transformar la experiencia de compra de viviendas en un entorno globalizado, combinando tecnologa de vanguardia con la toma de decisiones inteligente.

## <h1 style="font-family:Impact"><b><span style="color:blue;">Webs o APIS</span></b></h1>

En este proyecto, utilizaremos diferentes fuentes de informacin para analizar la economa y el mercado inmobiliario.

- Recopilaremos datos a travs del API del <span style="color:green;">Banco de la Reserva del Per</span> para obtener indicadores econmicos clave, como el ndice de precios al consumidor y la tasa de cambio.
- Adems, extraeremos datos de la pgina web de venta de viviendas <span style="color:green;">Urbania</span>  para analizar las tendencias y precios del mercado inmobiliario en tiempo real.

- Tambin utilizaremos la revista <span style="color:green;">The Economist</span>  para acceder a artculos y noticias relevantes sobre economa global y tendencias financieras. Estas fuentes nos permitirn obtener informacin actualizada y detallada para realizar un anlisis de nuestro proyecto.

Para este proyecto, se deja la puerta abierta para agregar ms fuentes de informacin en el transcurso del desarrollo. Si durante el desarrollo del proyecto tenemos la necesidad de investigar y analizar datos provenientes de otras pginas web o fuentes adicionales, estaremos abiertos a incorporarlas en nuestro anlisis.
