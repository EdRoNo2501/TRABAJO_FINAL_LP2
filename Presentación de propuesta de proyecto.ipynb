# <h1 style="font-family:Impact"><b><span style="color:blue;">Propósito</span></b></h1>

El proposito de este proyecto es proporcionar información relevante y un análisis que te ayuden a tomar decisiones informadas sobre la compra de una casa en el Perú. A través del análisis de indicadores económicos, como la inflación o el tipo de cambio, junto con el seguimiento de noticias y eventos económicos relacionados con el dólar, podrás evaluar el momento adecuado para realizar la compra. Además, podremos identificar el distrito más accesible en cuanto a precios de viviendas y realizar una comparación entre comprar en soles o en dólares, considerando los factores económicos relevantes.

# <h1 style="font-family:Impact"><b><span style="color:blue;">Objetivos</span></b></h1>


#### <h1 style="font-family:Impact"><b><span style="color:skyblue;">Objetivos Generales</span></b></h1>

- Se busca proporcionar recomendaciones basadas en el análisis realizado, con el objetivo de apoyar la toma de decisiones informadas en temas relacionados con la economía y el mercado inmobiliario en el Perú.


- Estudiar el mercado inmobiliario en Perú a través de la recopilación de datos de la página web de venta de viviendas.Para identificar oportunidades y desafíos en el sector inmobiliario.

- Identificar relaciones y posibles correlaciones entre los indicadores económicos, el mercado inmobiliario y los eventos económicos y políticos tanto a nivel nacional como internacional. Se buscará identificar patrones y tendencias que puedan influir en el desempeño económico y en el mercado inmobiliario del Perú.

#### <h1 style="font-family:Impact"><b><span style="color:skyblue;">Objetivos Especificos</span></b></h1>

- Realizar una comparacion de la economía peruana con respecto a la economía global y regional, utilizando información relevante y actualizada de la revista The Economist. Esto permitirá entender el contexto económico internacional y evaluar el impacto de los eventos globales en la economía del país.

- Analizar la evolución de la inflación en el Perú utilizando datos del índice de precios al consumidor proporcionados por el Banco de la Reserva del Perú. Esto permitirá entender las tendencias y variaciones en los precios de bienes y servicios, y su impacto en la economía.

- Analizar los precios de las viviendas de los distritos en Lima Metropolitana para obtener un panorama que nos sirva de guía al momento de realizar una compra.   

- Presentar los resultados del análisis de manera clara, utilizando visualizaciones de datos y gráficos que faciliten la comprensión de lo obtenido. 









 
## <h1 style="font-family:Impact"><b><span style="color:blue;">Webs o APIS</span></b></h1>

En este proyecto, utilizaremos diferentes fuentes de información para analizar la economía y el mercado inmobiliario.

- Recopilaremos datos a través del API del <span style="color:green;">Banco de la Reserva del Perú</span> para obtener indicadores económicos clave, como el índice de precios al consumidor y la tasa de cambio.
- Además, extraeremos datos de la página web de venta de viviendas <span style="color:green;">Properati</span>  para analizar las tendencias y precios del mercado inmobiliario en tiempo real.

- También utilizaremos la revista <span style="color:green;">The Economist</span>  para acceder a artículos y noticias relevantes sobre economía global y tendencias financieras. Estas fuentes nos permitirán obtener información actualizada y detallada para realizar un análisis de nuestro proyecto.

Para este proyecto, se deja la puerta abierta para agregar más fuentes de información en el transcurso del desarrollo. Si durante el desarrollo del proyecto tenemos la necesidad de investigar y analizar datos provenientes de otras páginas web o fuentes adicionales, estaremos abiertos a incorporarlas en nuestro análisis.

## <h1 style="font-family:Impact"><b><span style="color:blue;">Estrategias </span></b></h1>


Para realizar este proyecto se nesecita análisar sitios web de inmobiliarias, para obtener información sobre el precio de las casas en dólares y en moneda local; plataformas financieras, que ofrecen servicios de cambio de divisas en Perú, algunas de ellas podrían proporcionar información sobre el tipo de cambio actualizado. Y sitios web de noticias económicas, para mantenerse al tantode las noticias económicas y financieras en Perú, prestando especial atención a los informes sobre el tipo de cambio del dólar.

Utilizaremos técnicas de web scraping, para extraer datos y realizar un analisis de las páginas:
- Banco central de recerva del Perú: Proporciona api donde podemos encontrar el tipo de cambio nominal y la inflación que utilizaremos para el análisis. En el cual podremos ver como varia el valor del dolar en comparación de la monada del Perú(sol) en grafica y la inflasión en el pais en una tabla.
- The Econimist: Se hace un scraping la pagina con  las noticias relacionadas al dolar, para ver la situación en el que se encuentra usando BeautifulSoup.
- Properati: En esta pagina tambien encontramos los precios de las casas en dolares de lima, le haremos un scraping la pagina a las casas de Lima usaremos BeautifulSoup.
